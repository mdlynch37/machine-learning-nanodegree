<div>Separating line, or hyperplane</div>

<div>

*   Line that best separates data points on a scatter plot.
*   Maximizes <u>the margin</u>, or the distance to the nearest point relative to both classes.

<div>![](svm_images/5A6F8284-38F4-462D-B8E6-B65A63D56877.png)</div>

<div>(Selected point maximizes the margin, "most robust to maximization error")</div>

*   SVM first looks at classification, the margin maximization, to avoid any classification error.
*   If unavoidable classification error, SVM does the best it can. Truly ignores outliers.
*   <u>Maximizes marginal distance but can ignore outliers</u>.![](svm_images/13B67B54-C120-4FB2-BD34-67AA00B29887.png)

<div>**Important SVM parameters**</div>

<div>

<div><span style="font-family: 'Andale Mono';">sklearn.svm</span>.[SVC](http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC)</div>

</div>

<div><u>kernel : string, optional (default=’rbf’)</u></div>

*   Specifies the kernel type to be used in the algorithm. It must be one of ‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’ or a callable. If none is given, ‘rbf’ will be used. If a callable is given it is used to pre-compute the kernel matrix from data matrices; that matrix should be an array of shape (n_samples, n_samples).

<div>

<div>"kernel trick"</div>

*   Central to Machine Learning, uses manipulations of the x-axis to create non-linear decision boundaries from an algorithm that creates linear decision boundaries

<div>![](svm_images/EB08282D-C137-4666-BE4D-3EE67A9CEC72.png)</div>

</div>

<div><u>gamma : float, optional (default=’auto’)</u></div>

*   Kernel coefficient for ‘rbf’, ‘poly’ and ‘sigmoid’. If gamma is ‘auto’ then 1/n_features will be used instead.
*   Determines how linear the linear kernel is
*   Only for kernel=‘rbf'

<div><u>C : float, optional (default=1.0)</u></div>

*   Penalty parameter C of the error term.
*   Def: "Controls tradeoff between smooth decision boundary and classifying training points correctly."
*   > C parameter, > training points correctly classified.
