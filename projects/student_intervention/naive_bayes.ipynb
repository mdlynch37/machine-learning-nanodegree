<div><u>Why Naive?</u></div>

*   Because more information (that presumably could be available), the word order specifically, is not taken into account in the simple probability multiplication.
*   Used for textual analysis, naive because it does not take into account the word order. “Doesn’t understand the text, just word probabilities”.
*   Simple to run, efficient.
*   But can break in certain ways, like “chicago bulls” a phase that has meaning distinct of the words that comprise it.
*   Assumes independent features.
*   Words are <u>features</u>
*   Use theoretical understanding of algorithm to determine its suitability.
